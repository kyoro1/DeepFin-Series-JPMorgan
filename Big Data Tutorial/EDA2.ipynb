```
df = sqlContext.sql("SELECT * FROM asxhistoric2016")
```

```
display(df)
```

```
df.printSchema()
```

```
print(df.columns) 
```

```
from pyspark.sql.functions import col

tradingvolume_df = ((df  
  .withColumn("TradingVolume", df["Close"] * df["Volume"])
  ))
```

```
from pyspark.sql.functions import count
from pyspark.sql.functions import max, desc

display(tradingvolume_df
  .groupBy("Ticker")
  .agg(max("TradingVolume"))  
)
```
